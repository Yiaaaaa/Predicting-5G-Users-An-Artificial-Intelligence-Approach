```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd

#Import data
data = pd.read_csv('D:/train.csv')
data
```

```python
#Remove the first column 'id' and the last column 'target' from x_data
y_data = data['target'].values[0:]
x1 = data.drop('target', axis=1)
x_data=x1.drop('id',axis=1).values[0:]
print(x_data)
print(y_data)
```

```python
from sklearn import model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(x_data, y_data, test_size=0.2)
```

```python
#Create a logistic regression model object and train it using the fit method
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
#After training the samples, use the trained model to make predictions on the test set
y_pred = classifier.predict(x_test)
print("é¢„æµ‹ç»“æžœä¸ºï¼š",y_pred)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#Using the AUC standard to score the model's predicted results
from sklearn.metrics import roc_auc_score
x_pred=x_test
Y_pred=classifier.predict(x_pred)
y_true=y_test
roc_auc_score(y_true,y_pred)
```
